```python
import pyodbc
import pandas as pd
from datetime import datetime
```

```python
connection_string = 'DRIVER={ODBC Driver 17 for SQL Server};SERVER=REVISION-PC;DATABASE=CQ_BI_13_source_system;Trusted_connection=yes'
conn = pyodbc.connect(connection_string)
```

```python
df = pd.read_sql("SELECT * FROM supermarket_sales", conn)
df
```

```python
# Product ID
product_id_counter = df.groupby(["product_id"]).size().reset_index(name='Count')
product_id_counter
```

```python
# Invoice ID
invoice_id_counter = df.groupby(["invoice_id"]).size().reset_index(name='Count')
assert len(invoice_id_counter[invoice_id_counter['Count'] == 1]) > 0, "Has duplicated invoice id"
```

```python
# Invoice ID format
correct_format_mask = df['invoice_id'].str.match(r"^\d{3}-\d{2}-\d{4}$")
df_filtered = df[~correct_format_mask]
assert len(df_filtered) == 0, "Has invalid invoice id format"
```

```python
# Branch
branchs = df.groupby(["branch"]).size().reset_index(name='Count')
branchs
```

```python
# Customer Type
customer_types = df.groupby(["customer_type"]).size().reset_index(name='Count')
customer_types
```

```python
# Gender
genders = df.groupby(["gender"]).size().reset_index(name='Count')
genders
```

```python
# Quantity: quantity < 1 and value of quantity is interger
invalid_quantity_counter = df[(df["quantity"] < 1) | (~df["quantity"].astype(int).eq(df["quantity"]))]
assert len(invalid_quantity_counter) == 0, "Has invalid quantity values"
```

```python
# Tax 5%
invalid_tax_counter = df[df["tax_5"] < 0]
assert len(invalid_tax_counter) == 0, "Has invalid tax values"
```

```python
# Total
invalid_total_counter = df[df["total"] < 0]
assert len(invalid_total_counter) == 0, "Has invalid total values"
```

```python
# Date
purchage_date_out_of_range = df[~(df['date'].dt.year == 2019) & (df['date'].dt.month.between(1, 3))]
purchage_date_out_of_range
assert len(purchage_date_out_of_range) == 0, f"Has {len(purchage_date_out_of_range)} invalid rows"
```

```python
# Time
purchage_time_out_of_range = df[~df["time"].dt.hour.between(10, 21)]
assert len(purchage_time_out_of_range) == 0, f"Have {len(purchage_time_out_of_range)} invalid rows"
```

```python
# Payment
payment_types = df.groupby(["payment"]).size().reset_index(name='Count')
payment_types
```

```python
# cogs, gross margin percentage, gross income
invalid_cost_verifier = df[
    (df["cogs"] < 0)
    | (df["gross_margin_percentage"] < 0)
    | (df["gross_income"] < 0)
]
assert len(invalid_cost_verifier) == 0, "Has invalid rows"
```

```python
# Rating
rating_out_of_range = df[(df["rating"] > 10.) | (df["rating"] < 1.)]
assert len(rating_out_of_range) == 0, "Has invalid rows"
```

```python
df = pd.read_sql("SELECT * FROM product", conn)
df
```

```python
# Unit price
invalid_unit_price_counter = df[df["unit_price"] < 0]
assert len(invalid_unit_price_counter) == 0, "Has invalid unit_price values"
```

```python
# Product Line ID format
correct_format_mask = df['product_line'].str.match(r"^PD\d+$")
df_filtered = df[~correct_format_mask]
assert len(df_filtered) == 0, "Has invalid product line id format"
```
